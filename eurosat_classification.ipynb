```python
import tensorflow as tf
import keras
from tensorflow.keras import layers
import numpy as np
import os
import zipfile
import matplotlib.pyplot as plt # Import matplotlib for plotting

# --- Core Data Loading and Preprocessing ---

# Define the initial extracted directory and the nested zip file path
initial_extracted_dir = '/content' # Corrected: The zip is directly in /content
nested_zip_path = os.path.join(initial_extracted_dir, 'Group A -Eurosat Classification.zip') # Corrected: Use the correct zip file name and location

# Define the target directory for the final extraction
final_extract_to_dir = '/content/2750'

# Add a check to see if the zip file exists just before extraction
print(f"Checking for nested zip file at: {nested_zip_path}")
if os.path.exists(nested_zip_path):
    print(f"Nested zip file found at {nested_zip_path}. Attempting to extract its contents.")
    # Optional: List contents of the zip file to further confirm
    try:
        with zipfile.ZipFile(nested_zip_path, 'r') as zip_ref:
            print("Contents of the zip file:")
            for file_info in zip_ref.infolist():
                print(file_info.filename)
            zip_ref.extractall(final_extract_to_dir)
        print(f"Extraction of nested zip successful to {final_extract_to_dir}.")
        # Now set the data_dir to the location where the class folders should be
        # Corrected data_dir to point to the directory containing class folders
        data_dir = os.path.join(final_extract_to_dir, '2750')
        if not os.path.exists(data_dir) or not os.listdir(data_dir):
             print(f"Error: Final data directory {data_dir} not found or is empty after extraction.")
             data_dir = None
        else:
            print(f"Data directory {data_dir} found and appears to contain data.")

    except Exception as e:
        print(f"Error during extraction of nested zip: {e}")
        data_dir = None
else:
    print(f"Error: Nested zip file not found at {nested_zip_path}. Cannot proceed.")
    data_dir = None


# Proceed only if data_dir is valid
train_ds = None
val_ds = None
class_names = None

if data_dir and os.path.exists(data_dir):
    # Define image size and batch size
    img_height = 64
    img_width = 64
    batch_size = 32

    # Load the dataset using image_dataset_from_directory
    try:
        raw_ds = tf.keras.utils.image_dataset_from_directory(
          data_dir,
          labels='inferred',
          label_mode='int',
          image_size=(img_height, img_width),
          interpolation='nearest',
          batch_size=batch_size,
          shuffle=True
        )

        print("Dataset loaded successfully.")
        class_names = raw_ds.class_names
        print(f"Number of classes: {len(class_names)}")
        print(f"Class names: {class_names}")


        # Define preprocessing function (normalization)
        def preprocess_image(image, label):
            image = tf.cast(image, tf.float32) / 255.0
            return image, label

        # Apply normalization to the dataset
        processed_ds = raw_ds.map(preprocess_image)

        print("Dataset normalization applied.")

        # Split the dataset into training and validation sets
        dataset_size = tf.data.experimental.cardinality(processed_ds).numpy()
        if dataset_size > 0:
            train_size = int(0.8 * dataset_size)
            val_size = dataset_size - train_size

            if dataset_size == 1:
                 train_size = 1
                 val_size = 0
            elif train_size == 0 and dataset_size > 0:
                 train_size = 1
                 val_size = dataset_size - 1


            train_ds = processed_ds.take(train_size)
            val_ds = processed_ds.skip(train_size).take(val_size)

            print(f"Dataset split into {train_size} training batches and {val_size} validation batches.")

            # --- Apply Data Augmentation ---
            print("\nApplying Data Augmentation to Training Dataset...")
            data_augmentation = keras.Sequential(
                [
                    layers.RandomFlip("horizontal_and_vertical"),
                    layers.RandomRotation(0.2),
                    layers.RandomZoom(0.2),
                ]
            )

            # Apply data augmentation only to the training dataset
            train_ds = train_ds.map(lambda x, y: (data_augmentation(x, training=True), y),
                                    num_parallel_calls=tf.data.AUTOTUNE)

            print("Data augmentation applied to the training dataset.")
            # --- End Data Augmentation ---


            # Optional: Prefetching and caching for performance
            AUTOTUNE = tf.data.AUTOTUNE
            train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
            val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)

            print("Dataset cached and prefetched.")


        else:
            print("Dataset is empty after loading. Cannot split or preprocess.")

    except Exception as e:
        print(f"An error occurred during dataset loading or processing: {e}")
        print("Data loading and preprocessing failed.")
else:
    print("Data directory not found or extraction failed. Cannot proceed with model building and training.")


# --- Build the Model ---
model = None
if train_ds is not None and val_ds is not None and class_names is not None:
    try:
        num_classes = len(class_names)

        model = keras.Sequential([
            layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
            layers.MaxPooling2D((2, 2)),
            layers.Conv2D(64, (3, 3), activation='relu'),
            layers.MaxPooling2D((2, 2)),
            layers.Conv2D(128, (3, 3), activation='relu'),
            layers.MaxPooling2D((2, 2)),
            layers.Flatten(),
            layers.Dense(128, activation='relu'),
            layers.Dropout(0.5),
            layers.Dense(num_classes, activation='softmax')
        ])

        model.compile(optimizer='adam',
                      loss=tf.keras.losses.SparseCategoricalCrossentropy(),
                      metrics=['accuracy'])

        print("\nModel Architecture:")
        model.summary()

    except Exception as e:
        print(f"An error occurred during model building: {e}")
        model = None
        print("Model building failed.")

# --- Train the Model ---
history = None
if model is not None:
    try:
        epochs = 10
        print(f"\nTraining the model for {epochs} epochs...")
        history = model.fit(
            train_ds,
            validation_data=val_ds,
            epochs=epochs
        )
        print("Model training completed.")
    except Exception as e:
        print(f"An error occurred during model training: {e}")
        history = None
        print("Model training failed.")

# --- Evaluate the Model ---
evaluation_results = None
if model is not None and val_ds is not None:
    try:
        print("\nEvaluating the model on the validation dataset...")
        loss, accuracy = model.evaluate(val_ds)
        print(f"Validation Loss: {loss:.4f}")
        print(f"Validation Accuracy: {accuracy:.4f}")
        evaluation_results = {
            "loss": loss,
            "accuracy": accuracy
        }
    except Exception as e:
        print(f"An error occurred during model evaluation: {e}")
        print("Model evaluation failed.")

# --- Example Prediction ---
if model is not None and class_names is not None and val_ds is not None:
    print("\nExample Prediction:")
    try:
        for images, labels in val_ds.take(1):
            example_image_tensor = images[0]
            example_label_tensor = labels[0]

            img_array = np.expand_dims(example_image_tensor.numpy(), axis=0)
            true_label = example_label_tensor.numpy()
            print(f"Using an example image from the validation set with true label index: {true_label}")

        if 'img_array' in locals() and img_array is not None:
            predictions = model.predict(img_array)
            predicted_class_index = np.argmax(predictions[0])

            if 'class_names' in locals() and class_names is not None:
                predicted_class_name = class_names[predicted_class_index]
                print(f"Predicted class: {predicted_class_name}")
                if 'true_label' in locals():
                     true_class_name = class_names[true_label]
                     print(f"True class: {true_class_name}")
            else:
                print(f"Predicted class index: {predicted_class_index}")
                print("Class names not available to map index to name.")
        else:
            print("Could not create an image array for prediction.")
    except Exception as e:
        print(f"An error occurred during example prediction: {e}")
        print("Example prediction failed.")

# --- Visualize Model Performance ---
if history is not None:
    print("\nVisualizing Model Performance:")
    try:
        plt.figure(figsize=(12, 4))
        plt.subplot(1, 2, 1)
        plt.plot(history.history['accuracy'], label='Training Accuracy')
        plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
        plt.title('Training and Validation Accuracy')
        plt.xlabel('Epoch')
        plt.ylabel('Accuracy')
        plt.legend()
        plt.grid(True)

        plt.subplot(1, 2, 2)
        plt.plot(history.history['loss'], label='Training Loss')
        plt.plot(history.history['val_loss'], label='Validation Loss')
        plt.title('Training and Validation Loss')
        plt.xlabel('Epoch')
        plt.ylabel('Loss')
        plt.legend()
        plt.grid(True)

        plt.show()
        # display(plt.gcf()) # Removed display as it's not needed in a script

    except Exception as e:
        print(f"An error occurred during performance visualization: {e}")
        print("Performance visualization failed.")
```
